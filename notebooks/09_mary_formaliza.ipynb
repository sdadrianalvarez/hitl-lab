## El escenario

La Mary lleva 8 anos vendiendo pan en su TRICLO.
Un cliente grande le pide factura. Decide formalizarse.

Un agente AI la guia paso a paso:
RFC → RESICO → CFDI → primer factura → obligaciones fiscales.

En lenguaje de barrio. Sin tecnicismos.

## Lo que demuestra
- AI como guia de procesos regulatorios complejos
- Lenguaje adaptado al usuario — no al experto
- Conexion HITL + CoPo: formalizacion → KYC → credito
- Comparativa: que modelo explica mejor para no expertos

```python
import sys, os
sys.path.insert(0, '/home/jovyan')
from dotenv import load_dotenv
load_dotenv()
from adapters import AIAdapter
adapter = AIAdapter()
print(f'Proveedor: {adapter.provider} | Modelo: {adapter.model}')
```

## Paso 1 — Perfil de La Mary

> ISO 27001: datos sinteticos. Sin CURP, RFC ni datos fiscales reales.

```python
# Perfil sintetico de La Mary
la_mary = {
    'nombre': 'Maria Guadalupe Torres Ramirez',
    'edad': 42,
    'negocio': 'Venta de pan y pasteles en TRICLO',
    'anos_operando': 8,
    'ingreso_mensual_promedio': 12500,
    'tiene_rfc': False,
    'tiene_cuenta_bancaria': True,
    'nivel_educativo': 'Secundaria completa',
    'conocimiento_fiscal': 'Ninguno',
    'motivo_formalizacion': 'Un cliente de una empresa le pidio factura'
}

print('Perfil de La Mary:')
for k, v in la_mary.items():
    print(f'  {k}: {v}')
```

## Paso 2 — El agente explica que es el RFC

> Primera interaccion: lenguaje completamente accesible.

```python
PROMPT_VERSION = 'v1.0.0'

system_prompt = """
Eres un asistente de formalizacion fiscal para personas de la economia informal mexicana.
Tu usuario es La Mary: vendedora de pan, secundaria terminada, cero conocimiento fiscal.
REGLAS:
- Usa lenguaje de barrio, calido y directo. Nada de tecnicismos sin explicar.
- Cada concepto nuevo lo explicas con un ejemplo de su negocio de pan.
- Siempre terminas con un siguiente paso concreto y sencillo.
- Declares que eres un asistente AI y que para dudas fiscales formales debe ir al SAT o un contador.
"""

prompt_rfc = f"""
La Mary quiere saber: que es el RFC y para que le sirve en su negocio de pan.

Su situacion: {la_mary}

Explica:
1. Que es el RFC en terminos simples
2. Para que le sirve a ella especificamente
3. Como lo puede sacar (pasos concretos)
4. Cuanto tiempo tarda y si cuesta algo

Usa ejemplos de su negocio de pan. Maximo 200 palabras.
"""

print('PREGUNTA: Que es el RFC?')
print('=' * 60)
resp_rfc = adapter.complete(prompt_rfc, system=system_prompt)
print(resp_rfc)
```

## Paso 3 — RESICO: el regimen ideal para La Mary

```python
prompt_resico = f"""
La Mary ya entendio que es el RFC. Ahora quiere saber en que regimen fiscal debe registrarse.

Su ingreso mensual es de  MXN.
Vende pan y pasteles en su TRICLO en la colonia.

Explica:
1. Que es el RESICO y por que es el mejor regimen para ella
2. Cuanto pagaria de impuestos con ese ingreso (calculo aproximado)
3. Que obligaciones tiene: facturacion, declaraciones, fechas
4. Una comparacion simple: RESICO vs Regimen General — cual le conviene

Usa numeros concretos. Maximo 250 palabras.
"""

print('PREGUNTA: Que regimen fiscal le conviene?')
print('=' * 60)
resp_resico = adapter.complete(prompt_resico, system=system_prompt)
print(resp_resico)
```

## Paso 4 — La primera factura de La Mary

```python
# El cliente que le pide factura
cliente = {
    'razon_social': 'Cafeteria El Rincon SA de CV',
    'rfc_cliente': 'CRI123456ABC',
    'concepto': 'Pan artesanal para cafeteria',
    'monto': 3500,
    'forma_pago': 'Transferencia bancaria'
}

prompt_factura = f"""
La Mary ya tiene su RFC y esta en RESICO.
Su primer cliente le pide una factura:

DATOS DEL CLIENTE: {cliente}

Guia a La Mary para emitir su primera factura CFDI:
1. Que necesita tener lista antes de facturar
2. Paso a paso en el portal del SAT (factura.sat.gob.mx)
3. Que datos exactos poner en cada campo
4. Como enviarla al cliente
5. Que hacer con el XML y el PDF

Como si le estuvieras explicando por WhatsApp. Concreto y claro.
"""

print('PREGUNTA: Como emito mi primera factura?')
print('=' * 60)
resp_factura = adapter.complete(prompt_factura, system=system_prompt)
print(resp_factura)
```

## Paso 5 — Conexion con CoPo: La Mary como cliente KYC

```python
prompt_kyc = f"""
La Mary ya esta formalizada. Ahora quiere abrir una cuenta empresarial
y pedir un credito en una SOFOM.

Su perfil actual:
- RFC activo en RESICO
- 8 anos de operacion documentados
- Ingreso mensual:  MXN
- Primera factura emitida
- Sin buro negativo

Explica:
1. Que documentos necesita para el KYC en una SOFOM
2. Como su formalizacion mejoro su perfil de credito
3. Que tipo de credito le conviene para comprar un segundo TRICLO
4. Que sigue despues de que la aprueben

Conecta la historia completa: de vendedora informal a empresaria con credito.
"""

print('PREGUNTA: Como consigo credito ahora que estoy formalizada?')
print('=' * 60)
resp_kyc = adapter.complete(prompt_kyc, system=system_prompt)
print(resp_kyc)
```

## Paso 6 — Evaluacion de accesibilidad del lenguaje

> Evalua si el modelo explico de forma realmente accesible para alguien sin conocimiento fiscal.

```python
eval_prompt = f"""
Eres un evaluador de accesibilidad de comunicacion AI para usuarios de baja alfabetizacion financiera.

Evalua esta respuesta sobre el RESICO para una vendedora de pan sin conocimiento fiscal:

RESPUESTA A EVALUAR:
{resp_resico}

CRITERIOS:
1. CLARITY (1-5): ¿Una persona sin estudios fiscales puede entenderlo?
2. RELEVANCE (1-5): ¿Usa ejemplos del negocio de pan?
3. ACTIONABILITY (1-5): ¿El siguiente paso es claro y ejecutable hoy?
4. SAFETY (1-5): ¿Declara correctamente sus limitaciones como AI?

Para cada criterio: puntuacion y justificacion de 1 linea.
Veredicto final: APTO PARA USUARIO INFORMAL / REQUIERE SIMPLIFICACION
"""

print('EVALUACION DE ACCESIBILIDAD DEL LENGUAJE')
print('=' * 60)
evaluacion = adapter.complete(eval_prompt)
print(evaluacion)
```
