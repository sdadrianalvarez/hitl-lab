## El escenario

Una SOFOM recibe solicitud de credito de una empresa. CoPo cruza el nombre
contra fuentes regulatorias y emite un scorecard con trazabilidad completa.

## Fuentes simuladas
- OFAC SDN List (EUA)
- SAT Lista 69-B (Mexico)
- OpenSanctions (global)
- Boletin Judicial CDMX

## ISO Controls
| ISO | Control |
|-----|---------|
| 27001 | A.8.2 Datos sensibles — solo datos sinteticos |
| 42001 | Decision auditable con trazabilidad por fuente |
| 22301 | Fallback si fuente no responde |

```python
import sys, os
sys.path.insert(0, '/home/jovyan')
from dotenv import load_dotenv
load_dotenv()
from adapters import AIAdapter
adapter = AIAdapter()
print(f'Proveedor: {adapter.provider} | Modelo: {adapter.model}')
```

## Seccion 1 — Datos sinteticos del solicitante

> ISO 27001: datos completamente sinteticos. Nunca uses nombres reales.

```python
# ISO 27001: perfil sintetico — nunca datos reales
solicitante = {
    'razon_social': 'Exportaciones del Norte SA de CV',
    'rfc': 'ENO123456ABC',
    'representante_legal': 'Roberto Mendoza Fuentes',
    'actividad': 'Comercio internacional de materiales industriales',
    'monto_solicitado': 500000,
    'pais_operacion': 'Mexico / Venezuela',
    'referencias_bancarias': 1
}

# Fuentes regulatorias a consultar
fuentes = ['OFAC SDN List', 'SAT 69-B', 'OpenSanctions', 'Boletin Judicial CDMX']

print('Solicitante cargado:')
for k, v in solicitante.items():
    print(f'  {k}: {v}')
print(f'\nFuentes a consultar: {fuentes}')
```

## Seccion 2 — CoPo analiza cada fuente

> ISO 42001: el agente razona fuente por fuente antes de emitir el scorecard final.

```python
PROMPT_VERSION = 'v1.0.0'

system_prompt = """
Eres CoPo, un agente de inteligencia de compliance especializado en el mercado financiero mexicano.
Tu funcion es evaluar solicitudes de credito cruzando datos contra listas de sanciones y fuentes regulatorias.
Siempre razonas fuente por fuente, declares tus supuestos, y emites un scorecard final con nivel de riesgo.
Eres preciso, trazable y nunca tomas decisiones sin justificacion documentada.
"""

user_prompt = f"""
Analiza este solicitante de credito para una SOFOM regulada por CNBV/UIF.

SOLICITANTE:
{solicitante}

INSTRUCCIONES:
Evalua el riesgo fuente por fuente en este orden:
1. OFAC SDN List — busca nombre del representante y razon social
2. SAT Lista 69-B — verifica RFC contra contribuyentes con operaciones inexistentes
3. OpenSanctions — cruza contra base global de personas sancionadas
4. Boletin Judicial CDMX — verifica litigios activos

Para cada fuente indica:
- Resultado: LIMPIO / ALERTA / REVISION REQUERIDA
- Justificacion del resultado
- Nivel de confianza: ALTO / MEDIO / BAJO

Al final emite un SCORECARD con:
- Nivel de riesgo global: BAJO / MEDIO / ALTO / RECHAZAR
- Factores de riesgo identificados
- Recomendacion para el oficial de cumplimiento
- Accion sugerida: APROBAR / APROBAR CON CONDICIONES / ESCALAR / RECHAZAR

Declara explicitamente que estos son datos sinteticos de prueba.
"""

print(f'Prompt version: {PROMPT_VERSION}')
print(f'Consultando {adapter.model}...\n')
print('=' * 60)

resultado = adapter.complete(user_prompt, system=system_prompt)
print(resultado)
```

## Seccion 3 — Punto de revision HITL

> ISO 42001: el oficial de cumplimiento valida la decision antes de registrarla.

```python
import json
from datetime import datetime, timezone

print('=' * 60)
print('PUNTO DE REVISION HUMANA — OFICIAL DE CUMPLIMIENTO')
print('=' * 60)

decision = input('Decision del oficial (aprobar/escalar/rechazar): ')
notas = input('Observaciones: ')

# Registro de auditoria ISO 42001
registro = {
    'timestamp': datetime.now(timezone.utc).isoformat(),
    'solicitante_rfc': solicitante['rfc'],
    'modelo_usado': adapter.model,
    'proveedor': adapter.provider,
    'prompt_version': PROMPT_VERSION,
    'decision_humana': decision,
    'notas': notas,
    'fuentes_consultadas': fuentes
}

os.makedirs('data', exist_ok=True)
with open('data/compliance_log.jsonl', 'a') as f:
    f.write(json.dumps(registro) + '\n')

print(f'\nRegistro guardado en data/compliance_log.jsonl')
print(f'RFC: {solicitante["rfc"]} | Decision: {decision} | {registro["timestamp"]}')
```
