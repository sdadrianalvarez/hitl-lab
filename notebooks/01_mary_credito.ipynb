## El problema

La Mary lleva 8 años vendiendo pan en su TRICLO en la Colonia. Quiere pedir un microcrédito
de $15,000 MXN para comprar un segundo TRICLO y expandirse a otra esquina.

Un agente AI analiza su caso y decide si la aprueba — explicando cada paso.

---

## ISO Controls cubiertos

| ISO | Control | Cómo se aplica aquí |
|-----|---------|---------------------|
| **27001** | A.8.2 Clasificación de información | Datos de La Mary son sintéticos — nunca datos reales |
| **42001** | 6.1 Transparencia | El agente explica su razonamiento paso a paso |
| **22301** | 8.4 Continuidad | Si el modelo falla, hay respuesta de fallback documentada |

## Sección 1 — Configuración del ambiente

````python
import sys, os
sys.path.insert(0, '/home/jovyan')  # asegura que adapters/ sea importable

from dotenv import load_dotenv
load_dotenv()

from adapters import AIAdapter

# ── Cambia PROVIDER en .env para cambiar de nube ──
adapter = AIAdapter()
print(f'✅ Proveedor activo: {adapter.provider}')
print(f'✅ Modelo: {adapter.model}')
````

## Sección 2 — Datos sintéticos de La Mary

> **ISO 27001**: estos datos son completamente sintéticos. No uses datos reales de clientes.

````python
# ── Perfil sintético — La Mary (ISO 27001: datos ficticios) ──
la_mary = {
    'nombre': 'María Guadalupe Torres Ramírez',
    'negocio': 'Venta de pan y pasteles - TRICLO',
    'anos_operando': 8,
    'ingreso_mensual_promedio': 12500,  # MXN
    'gastos_mensuales': 7800,
    'ahorro_mensual': 4700,
    'monto_solicitado': 15000,
    'historial_credito': 'Sin buró registrado',
    'garantia': 'Ninguna',
    'referencias': 2,
    'colonia': 'Colonia HITL, Querétaro'
}

print('📋 Perfil de La Mary cargado:')
for k, v in la_mary.items():
    print(f'  {k}: {v}')
````

## Sección 3 — El agente analiza el crédito

> **ISO 42001**: el prompt está documentado y versionado aquí. El output es revisado por humano antes de usarse.

````python
# ── ISO 42001: prompt documentado y versionado ──
PROMPT_VERSION = 'v1.0.0'

system_prompt = """
Eres un analista de microcréditos especializado en economía informal mexicana.
Tu objetivo es evaluar solicitudes de crédito de manera justa, transparente y accesible.
Siempre explicas tu razonamiento paso a paso.
Usas lenguaje claro, sin tecnicismos innecesarios.
"""

user_prompt = f"""
Analiza esta solicitud de microcrédito usando razonamiento paso a paso.

SOLICITANTE:
{la_mary}

INSTRUCCIONES:
1. Calcula el flujo de caja mensual disponible
2. Evalúa la capacidad de pago considerando una tasa del 3% mensual
3. Identifica factores de riesgo y factores positivos
4. Emite una recomendación: APROBAR / APROBAR CON CONDICIONES / RECHAZAR
5. Explica la decisión en lenguaje que La Mary pueda entender

Muestra tu razonamiento en cada paso antes de concluir.
"""

print(f'📝 Prompt versión: {PROMPT_VERSION}')
print(f'📡 Enviando al modelo {adapter.model}...\n')

respuesta = adapter.complete(user_prompt, system=system_prompt)
print(respuesta)
````

## Sección 4 — Revisión humana (HITL)

> **ISO 42001**: el output del modelo SIEMPRE pasa por revisión humana antes de comunicarse al cliente.

````python
# ── HITL: el analista humano valida la decisión ──
print('=' * 60)
print('PUNTO DE REVISIÓN HUMANA — HITL')
print('=' * 60)
print('Un analista humano debe revisar la recomendación anterior.')
print()

decision_humana = input('¿Confirmas la recomendación del modelo? (si/no/modificar): ')
notas = input('Notas del analista (opcional): ')

print(f'\n✅ Decisión registrada: {decision_humana}')
if notas:
    print(f'📝 Notas: {notas}')
````
