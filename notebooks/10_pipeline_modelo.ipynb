## Paso 1 — Recoleccion de datos

> ISO 27001: datos sinteticos generados programaticamente. Nunca datos reales de clientes.

```python
import json, random
from datetime import datetime, timedelta

# Genera dataset sintetico de ventas de La Mary
random.seed(42)
productos = ['Concha', 'Cuerno', 'Polvoron', 'Dona', 'Baguette']
precios = {'Concha': 18, 'Cuerno': 15, 'Polvoron': 12, 'Dona': 20, 'Baguette': 35}

ventas = []
fecha_base = datetime(2026, 1, 1)

for i in range(90):  # 90 dias de ventas
    fecha = fecha_base + timedelta(days=i)
    es_fin_semana = fecha.weekday() >= 5
    n_ventas = random.randint(15, 30) if es_fin_semana else random.randint(8, 18)
    for _ in range(n_ventas):
        producto = random.choice(productos)
        cantidad = random.randint(1, 5)
        ventas.append({
            'fecha': fecha.strftime('%Y-%m-%d'),
            'dia_semana': fecha.strftime('%A'),
            'producto': producto,
            'cantidad': cantidad,
            'precio_unitario': precios[producto],
            'total': precios[producto] * cantidad
        })

os.makedirs('data/synthetic', exist_ok=True)
with open('data/synthetic/ventas_mary.json', 'w') as f:
    json.dump(ventas, f, ensure_ascii=False, indent=2)

print(f'Dataset generado: {len(ventas)} registros de venta')
print(f'Ejemplo: {ventas[0]}')
```

## Paso 2 — Limpieza y resumen estadistico

```python
import pandas as pd

df = pd.DataFrame(ventas)

# Limpieza basica
df['fecha'] = pd.to_datetime(df['fecha'])
df = df.dropna()
df = df[df['total'] > 0]

# Resumen
resumen = {
    'total_registros': len(df),
    'ingreso_total_MXN': int(df['total'].sum()),
    'ingreso_promedio_diario': round(df.groupby('fecha')['total'].sum().mean(), 2),
    'producto_mas_vendido': df.groupby('producto')['cantidad'].sum().idxmax(),
    'dia_mas_rentable': df.groupby('dia_semana')['total'].sum().idxmax(),
    'periodo': f"{df['fecha'].min().date()} a {df['fecha'].max().date()}"
}

print('Resumen estadistico del negocio de La Mary:')
for k, v in resumen.items():
    print(f'  {k}: {v}')
```

## Paso 3 — Construccion del contexto RAG

> RAG = Retrieval Augmented Generation. En lugar de fine-tuning, le damos contexto real al modelo en el prompt.

```python
# Construye el contexto RAG con los datos reales del negocio
contexto_rag = f"""
DATOS DEL NEGOCIO DE LA MARY (fuente: sistema de ventas TRICLO)
Periodo: {resumen['periodo']}
Total de transacciones: {resumen['total_registros']}
Ingreso total:  MXN
Ingreso promedio diario:  MXN
Producto estrella: {resumen['producto_mas_vendido']}
Dia mas rentable: {resumen['dia_mas_rentable']}

VENTAS POR PRODUCTO:
{df.groupby('producto')[['cantidad','total']].sum().to_string()}

VENTAS POR DIA DE SEMANA:
{df.groupby('dia_semana')['total'].sum().sort_values(ascending=False).to_string()}
"""

print('Contexto RAG construido:')
print(contexto_rag)
```

## Paso 4 — Inferencia con RAG

> El modelo responde preguntas de negocio usando los datos reales de La Mary como contexto.

```python
PROMPT_VERSION = 'v1.0.0'

system_prompt = """
Eres un consultor de negocios especializado en economia informal mexicana.
Usas UNICAMENTE los datos proporcionados para responder.
Si la informacion no esta en los datos, lo dices explicitamente.
Das recomendaciones practicas y accionables en lenguaje simple.
"""

# Pregunta de negocio
pregunta = '¿Cual es mi mejor dia para vender y que producto deberia promocionar mas? Dame 3 recomendaciones concretas.'

user_prompt = f"""
CONTEXTO DEL NEGOCIO:
{contexto_rag}

PREGUNTA DE LA MARY:
{pregunta}

Responde basandote UNICAMENTE en los datos anteriores.
Indica que datos especificos usaste para cada recomendacion.
"""

print(f'Pregunta: {pregunta}')
print(f'Modelo: {adapter.model} | Version prompt: {PROMPT_VERSION}')
print('=' * 60)

respuesta_rag = adapter.complete(user_prompt, system=system_prompt)
print(respuesta_rag)
```

## Paso 5 — Evaluacion de calidad (como Outlier)

> Este paso replica exactamente lo que hacen los raters de Outlier: evaluar la calidad de una respuesta AI con criterios objetivos.

```python
eval_prompt = f"""
Eres un evaluador experto de respuestas de AI (similar a un rater de Outlier AI).

Evalua esta respuesta usando los 4 criterios estandar:

PREGUNTA ORIGINAL: {pregunta}

CONTEXTO DISPONIBLE: {contexto_rag[:500]}...

RESPUESTA A EVALUAR:
{respuesta_rag}

CRITERIOS DE EVALUACION:
1. ACCURACY (1-5): ¿La respuesta usa correctamente los datos del contexto?
2. COMPLETENESS (1-5): ¿Responde todo lo que se pregunto?
3. NEUTRALITY (1-5): ¿Evita sesgos y declaraciones sin evidencia en los datos?
4. PRACTICAL UTILITY (1-5): ¿Las recomendaciones son accionables para La Mary?

Para cada criterio: puntuacion, justificacion de 2 lineas.
Al final: score total y veredicto en una linea.
"""

print('EVALUACION DE CALIDAD DE LA RESPUESTA AI')
print('=' * 60)
evaluacion = adapter.complete(eval_prompt)
print(evaluacion)
```
